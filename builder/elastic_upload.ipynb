```python
# Định nghĩa mapping cho index phù hợp với cấu trúc dữ liệu của bạn
index_name = "utehy_v1"
mappings = {
    "properties": {
        "title": {"type": "text", "analyzer": "standard", "fields": {"keyword": {"type": "keyword"}}},
        "content": {"type": "text", "analyzer": "standard"},
        "metadata": {
            "type": "nested",
            "properties": {
                "source": {"type": "text", "analyzer": "standard"},
            }
        }
    }
}

# Tạo index với mapping
if not es.indices.exists(index=index_name):
    es.indices.create(
        index=index_name,
        mappings=mappings,
        settings={
            "number_of_shards": 1,
            "number_of_replicas": 0,
            "analysis": {
                "analyzer": {
                    "vietnamese_analyzer": {
                        "type": "custom",
                        "tokenizer": "standard",
                        "filter": ["lowercase", "asciifolding"]
                    }
                }
            }
        }
    )
    print(f"Index '{index_name}' đã được tạo thành công!")
else:
    print(f"Index '{index_name}' đã tồn tại!")
```

```python
from elasticsearch.helpers import bulk


# Hàm để chuẩn bị dữ liệu cho bulk insert
def generate_bulk_data(data_items, index_name):
    for i, item in enumerate(data_items):
        yield {
            "_index": index_name,
            "_id": i + 1,  # ID tự tạo bắt đầu từ 1
            "_source": item
        }

# Upload dữ liệu với bulk API
def upload_custom_data(data_items, index_name="test"):
    try:
        success, failed = bulk(
            es,
            generate_bulk_data(data_items, index_name),
            refresh=True
        )
        print(f"Đã upload thành công {success} items. Thất bại: {len(failed) if failed else 0}")
        return success
    except Exception as e:
        print(f"Lỗi khi upload dữ liệu: {e}")
        return 0

# Gọi hàm để upload
upload_custom_data(data)

# Kiểm tra số lượng documents
count = es.count(index="test")
print(f"Tổng số items trong index: {count['count']}")
```

```python
class ElasticSearcher:
    def __init__(self, es_client, index_name="test"):
        self.es = es_client
        self.index_name = index_name

    def search(self, query_text, size=3):
        search_query = {
            "query": {
                "bool": {
                    "should": [
                        {
                            "match": {
                                "title": {"query": query_text, "boost": 1.0}
                            }
                        },
                        {
                            "match": {
                                "content": {"query": query_text, "boost": 3.0}
                            }
                        }
                    ],
                    "minimum_should_match": 1
                }
            }
        }

        try:
            response = self.es.search(
                index=self.index_name,
                body=search_query,
                size=size
            )
            
            results = []
            for hit in response['hits']['hits']:
                result = {
                    'score': hit['_score'],
                    'title': hit['_source']['title'],
                    'content': hit['_source']['content'],
                    'source': hit['_source']['metadata']['source']
                }
                results.append(result)
            
            return results
            
        except Exception as e:
            print(f"Search error: {e}")
            return []

# Usage example:
searcher = ElasticSearcher(es)
```

```python
# Example search
results = searcher.search("điểm sàn xét tài năng")
for i, result in enumerate(results, 3):
    print(f"\nResult #{i}")
    print(f"Title: {result['title']}")
    print(f"Content: {result['content']}")
    print(f"Source: {result['source']}")
    print("-" * 50)
```
